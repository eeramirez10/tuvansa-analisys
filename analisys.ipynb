```python
df["FechaEmision"] = pd.to_datetime(df["FechaEmision"])
```

```python
df["Month"] =df["FechaEmision"].dt.month
df["Year"] = df["FechaEmision"].dt.year
```

```python

empresasGroupByNameEmisor = df.groupby(by="RfcEmisor").agg({"NombreRazonSocialEmisor":"first"})
empresasGroupByNameEmisor = empresasGroupByNameEmisor.where(empresasGroupByNameEmisor["NombreRazonSocialEmisor"] != "NombreRazonSocialEmisor").dropna().reset_index()
# empresasName = empresasGroupByNameEmisor["NombreRazonSocialEmisor"].reset_index()
# empresasName

empresaEmisora = df[(df["RfcEmisor"] == "LAV751101B63") & (df["Year"] == 2023)]


top50Emisores = empresaEmisora.groupby(by="RfcReceptor").agg({"NombreRazonSocialReceptor":"first", "Total":"sum"}).sort_values("Total", ascending=False).head(50).reset_index()

empresaEmisora.groupby(["Month","Year"]).agg({"Total":"sum"}).reset_index()
```
